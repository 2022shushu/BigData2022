```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

#https://smtownandstore.com/product/celeb_nct.html?cate_no=53
base_url = "https://smtownandstore.com"
nct_total_url = "/product/celeb_nct.html?cate_no=53"
base_url + nct_total_url
```

```python
response = requests.get(base_url+nct_total_url, headers={"User-Agent": "Mozilla/5.0"})                                        
```

```python
soup = BeautifulSoup(response.content, "lxml")
soup
```

```python
nct_info = soup.findAll('div', {'class':'description'})
nct_info[0]
```

```python
nct_url = nct_info[0].find('a').get('href')
print(nct_url)
```

```python
name = nct_info[0].find('div', {'class':'name'}).text
print(name)
```

```python
price = nct_info[0].find('li', {'class':'xans-record-'}).text
print(price)
```

```python
name_list = []
price_list = []
url_list = []
```

```python
for nct in nct_info:
    name = nct.find('div', {'class':'name'}).text
    price = nct.find('li', {'class':'xans-record-'}).text
    url = nct.find('a').get('href')
    print("이름: {}, 가격 : {}".format(name, price))
    name_list.append(name)
    price_list.append(price)
    url_list.append(url)
```

```python
search = '상품명 : '
for i, word in enumerate(name_list):
    if search in word: 
        print('>> modify: ' + word)
        name_list[i] = word.strip(search)
        
search2 = '\n판매가 : '
for i, word in enumerate(price_list):
    if search2 in word: 
        print('>> modify: ' + word)
        price_list[i] = word.strip(search2)

search3 = '\n할인판매가 : '
for i, word in enumerate(price_list):
    if search3 in word: 
        print('>> modify: ' + word)
        price_list[i] = word.strip(search3)
        
search4 = '19%'
for i, word in enumerate(price_list):
    if search4 in word: 
        print('>> modify: ' + word)
        price_list[i] = word.strip(search4)

               
search5 = '10%'
for i, word in enumerate(price_list):
    if search5 in word: 
        print('>> modify: ' + word)
        price_list[i] = word.strip(search5)
```

```python
df = pd.DataFrame({"이름":name_list, "가격" : price_list, "url" : url_list})
df.head()
```
